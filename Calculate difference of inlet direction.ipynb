```python
import numpy as np

def unit_vector(vector):
    """Returns the unit vector of the input vector."""
    return vector / np.linalg.norm(vector)

def angle_between(v1, v2):
    """
    Returns the angle in radians between vectors 'v1' and 'v2'.
    """
    v1_u = unit_vector(v1)
    v2_u = unit_vector(v2)
    return np.arccos(np.clip(np.dot(v1_u, v2_u), -1.0, 1.0))

# Example usage:
# Suppose you have the inlet wind direction vector (D) and the local terrain normal vector (N):
inlet_wind_direction_degrees = 0
inlet_wind_direction_radians = np.radians(inlet_wind_direction_degrees)
D = (np.cos(inlet_wind_direction_radians), np.sin(inlet_wind_direction_radians))
```
